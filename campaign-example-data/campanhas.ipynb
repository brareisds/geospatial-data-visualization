```python
import pandas as pd

# Caminho para o arquivo CSV
file_path = 'microrregiao.csv'

# Carregar o arquivo CSV em um DataFrame do pandas
df = pd.read_csv(file_path)

# Exibir as primeiras linhas do DataFrame para verificar se os dados foram carregados corretamente
df.head()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely import wkt

# Converter a coluna de geometria de WKT para objetos geomtricos
df['geometria'] = df['geometria'].apply(wkt.loads)

# Converter o DataFrame do pandas para um GeoDataFrame do geopandas
gdf = gpd.GeoDataFrame(df, geometry='geometria')

# Plotar o mapa
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightblue', edgecolor='black')
plt.title('Mapa das Microrregies')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
